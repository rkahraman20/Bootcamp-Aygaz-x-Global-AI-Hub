## Listeler

- İndekslenebilirdirler.
- Birden fazla değer saklayabiliriz.
- Farklı veri tiplerini içerebilirler.

```python
a = ['a', 5, True]

print(a)
print(type(a))
print(type(a[0]))
print(type(a[1]))
print(type(a[2]))
```

```python
b = list()
c = []

print(type(b))
print(type(c))
```

```python
isimler = ['ali', 'veli', 'selami']
print(isimler)

isimler.append('ridvan')
print(isimler)
```

```python
isimler.insert(0, 'ayhan')
print(isimler)
```

```python
isimler.insert(2, 'gökhan')
print(isimler)
```

```python
isimler.extend('cenk')
print(isimler)
```

```python
isimler.extend(['azra', 'elif', 'zehra'])
print(isimler)
```

```python
isimler.pop()
print(isimler)
```

```python
isimler.remove('c')
print(isimler)
```

```python
isimler.clear()
print(isimler)
```

## Vücut Kitle İndeksi

```python
kisiler = []
isim = str(input('Bir isim giriniz: '))
kilo = float(input('Kilonuzu giriniz: '))
boy = float(input('Boyunuzu giriniz: '))
vki = round(kilo/((boy/100)**2))
print(vki)
kisi = {isim: vki}
kisiler.append(kisi)
print(type(kisiler))
print(kisiler)
```
